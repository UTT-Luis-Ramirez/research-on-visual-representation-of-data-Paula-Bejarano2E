###1. **Histograma**

---


**¿Que es?**

Es un grafico que muestra como se distribuyen los datos, dividiendolos en intervalos(o bins).Representa la frecuencia de datos dentro de cada intervalo.


---


**¿Como se usa?**

Se utiliza para analizar la distribucion de un conjunto de datos.Por ejemplo, saber cuantos estudiantes sacaron calificaciones en ciertos rangos.


---


**Contexto**

Util para entender patrones en datos, como en estadisticas, negocios, y ciencias.Por ejemplo, analizar las edades de un grupo para saber donde se concentra la mayoria.



###CÓDIGO Y GRÁFICA

---



```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(loc=50, scale=15, size=100)
plt.hist(data, bins=10, color='skyblue', edgecolor='black')
plt.title("Histograma")
plt.xlabel("Intervalos de valores")
plt.ylabel("Frecuencia")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

###2.**DIAGRAMA DE TALLO Y HOJA**

---



**¿Que es?**
Es un metodo que organiza datos numericos dividiendolos en dos partes: el tallo(parte principal) y la hoja(el digito final).Es como una tabla visual.

---



**¿Como se usa?**
Para observar la forma y distribucion de los datos de manera rapida, sin usar tecnologia avanzada.Es comun en estadisticas.




---


**Contexto**
Ideal para pequeños conjuntos de datos, como analizar calificaciones de un salon o puntuaciones en un juego.

###3.**DIAGRAMA DE CAJA (BOXPLOT)**


---



**¿Que es?**
Es un grafico que muestro la **dispersion de los datos**(mediana,cuartiles y valores atipicos).Ayuda a visualizar la variabilidad y detectar datos extremos.

---



**¿Como se usa?**
Para resumir datos y compararlos entre grupos.Por ejemplo, comparar ingresos en diferentes ciudades o calificaciones de distintas materias.




---


**Contexto**
Se usa en analisis estadistico, insvestigacion cientifica y hasta en negocios para estudiar tendencias.

######CÓDIGO Y GRÁFICA

---



```python
data=[45,47,49,50,52,53,55,60,65,70,75,100]
plt.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title("Diagrama de caja")
plt.xlabel("Valores")
plt.show()
```
